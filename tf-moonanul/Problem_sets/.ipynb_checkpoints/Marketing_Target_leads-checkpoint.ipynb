Check current occupation for specalization NAN value of  to see if we could get more information on missing value

```python
# Specialization with NAN and its corresponding current occupation
Leads[Leads['Specialization'].isnull()] ['What is your current occupation'].value_counts(dropna=False)
```

- When 'Specialization' is with NAN  and its corresponding 'current occupation' is also NAN then we can replace those with Other.
- For 'Specialization' with NAN  and has value in 'current occupation' then we can find mode in such a way that- Specialization which has higest count for that 'current occupation'.

```python
# Specialization with Select and its corresponding current occupation
Leads[Leads['Specialization'] == "Select"] ['What is your current occupation'].value_counts(dropna=False)
```

- Unemployed & Student arent employed and hence they dont have work specalization. They can be gropued under new name 'entry level'.
- When 'Specialization' is with Select  and its corresponding 'current occupation' is also NAN then we can replace those with Other.
- For 'Specialization' with Select  and has value in 'current occupation' then we can find mode in such a way that- filter data for that occupation and then find mode for Specialization.

```python
# Replacing NAN value of Specialization which has some value in current occupation. Filter data for that occupation and then find mode for Specialization.
Leads['Specialization'].fillna(value=Leads['What is your current occupation'].apply(lambda x : np.nan if pd.isnull(x) else  Leads[Leads['What is your current occupation'] == x ] ['Specialization'].mode()[0] ),inplace=True)
```

```python
# Replacing Select value of Specialization which has some value in current occupation. Filter data for that occupation and then find mode for Specialization
def Imp_Select_Spz(x):
    y = ''
    if x['Specialization'] == "Select" and pd.isnull(x['What is your current occupation']):
        y = 'Other'
    elif  x['Specialization'] == "Select":
        y = Leads[Leads['What is your current occupation'] == x['What is your current occupation'] ] ['Specialization'].mode()[0]
    else:
        y = x['Specialization']
    return y

Leads['Specialization'] = Leads.apply(lambda x :  Imp_Select_Spz(x)  , axis = 1 )
```

```python
# Rest of 'select options' which is unemployed and student can be grouped as 'Entery Level' 
Leads['Specialization'] = Leads['Specialization'].apply(lambda x: 'Entry Level' if (x == "Select")  else x )
```

```python
#Rest of Specialization NAN value  which is NAN and also its occupation data is NAN can be grouped under 'Others'
Leads['Specialization'].fillna(value= 'Other',inplace=True)
```

```python
plt.figure(figsize=(15,5))
runt=sns.countplot(Leads.Specialization)
runt.set_xticklabels(runt.get_xticklabels(),rotation=90)
runt.axes.set_title("Specialization after NaNs fill", fontsize = 15)
plt.show()
```

As few category item count are very low and to avoid model bias, we could club catgory as mentioned below
- Finance = ['Finance Management', 'Banking, Investment And Insurance' ]
- ECOMMERCE = ['E-COMMERCE', 'Retail Management', 'Supply Chain Management','E-Business',. 'Services Excellence'  ]
- Business/Administration  = ['Business Administration', 'International Business', 'Travel and Tourism','Rural and Agribusiness','Healthcare Management','Hospitality Management'  ]
- Media/Marketing  = ['Marketing Management', 'Media and Advertising' ]

```python
# Grouping all Finance related Specialization
Finance = ['Finance Management', 'Banking, Investment And Insurance' ]
Leads['Specialization'] = Leads['Specialization'].apply( lambda x: 'Finance' if x in Finance else x)

# Grouping all ECOMMERCE related Specialization
ECOMMERCE = ['E-COMMERCE', 'Retail Management', 'Supply Chain Management','E-Business'  ]
Leads['Specialization'] = Leads['Specialization'].apply( lambda x: 'E-COMMERCE' if x in ECOMMERCE else x)

# Grouping all IT PROJECT related Specialization
IT_PROJECT = ['IT Projects Management', 'Operations Management', 'Services Excellence' ]
Leads['Specialization'] = Leads['Specialization'].apply( lambda x: 'IT PROJECT' if x in IT_PROJECT else x)

# Grouping all Business/Administration related Specialization
Business_Administration  = ['Business Administration', 'International Business', 'Travel and Tourism','Rural and Agribusiness','Healthcare Management','Hospitality Management'  ]
Leads['Specialization'] = Leads['Specialization'].apply( lambda x: 'Business/Administration' if x in Business_Administration else x)

# Grouping all Media/Marketing related Specialization
Marketing  = ['Marketing Management', 'Media and Advertising' ]
Leads['Specialization'] = Leads['Specialization'].apply( lambda x: 'Media/Marketing' if x in Marketing else x)
```

- current occupation which NAN and have some in 'Specialization' then we can find mode in such a way that- filter data for that Specialization and then find mode for current occupation.

```python
# Replacing NAN of current occupation which has some value in Specialization. Filter data for that Specialization  and then find mode for occupation
def Imp_Curr_Occ(x):
    y = ''
    if x['Specialization'] == "Other" and pd.isnull(x['What is your current occupation']):
        y = 'Other'
    elif  pd.isnull(x['What is your current occupation']):
        y = Leads[Leads['Specialization'] == x['Specialization'] ] ['What is your current occupation'].mode()[0]
    else:
        y = x['What is your current occupation']
    return y

Leads['What is your current occupation'] = Leads.apply(lambda x :  Imp_Curr_Occ(x)  , axis = 1 )
```

```python
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(4, 4))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic for log reg#6')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
fpr, tpr, thresholds = metrics.roc_curve( y_train_pred_final.Converted, y_train_pred_final.Converted_prob, drop_intermediate = False )
```
